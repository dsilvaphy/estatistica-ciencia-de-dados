# Média Móvel

```python
# Importando os pacotes que iremos utilizar
import pandas as pd
import numpy as np
import random
```

```python
# Criando uma lista vazia
dados = []

# Atribuindo um range de datas para uma variavel periodo_data
periodo_data = pd.date_range("01/01/2023", "01/07/2023")

# Loop for para inserir os dados na lista chamada "data"
for i in range(len(periodo_data)):

    vendas_dinheiro = random.randint(1, 10)
    vendas_cartao = random.randint(1, 10)

    tipoDia = "Dia da Semana"

    if periodo_data[i].weekday()>4:
        tipoDia = "Final de Semana"

    dados.append([str(periodo_data[i].date()), tipoDia, "Dinheiro", vendas_dinheiro])
    dados.append([str(periodo_data[i].date()), tipoDia, "Cartão", vendas_cartao])

dados = pd.DataFrame(dados, columns=["Data", "Tipo_Dia", "Forma_Pagamento", "Qtd_Vendas"])
dados["Data"] = pd.to_datetime(dados["Data"])
     
```

```python
dados
```

```python
# Cria uma coluna com a soma de uma janela de 2 linhas
dados['soma_2dias'] = dados.rolling(window=2).Qtd_Vendas.sum()
```

```python
# Cria uma coluna com a soma de uma janela de 3 linhas
dados['soma_3dias'] = dados.rolling(window=3).Qtd_Vendas.sum()
```

```python
# Cria uma coluna com a média de uma janela de 2 linhas
dados['media_2dias'] = dados.rolling(window=2).Qtd_Vendas.mean()
```

```python
# Cria uma coluna com a média de uma janela de 3 linhas
dados['media_3dias'] = dados.rolling(window=3).Qtd_Vendas.mean()
```

```python
# Cria uma coluna com a média de uma janela de 7 dias
dados['med_movel_7dias'] = dados.rolling(window='7d', on='Data').Qtd_Vendas.mean()
```
