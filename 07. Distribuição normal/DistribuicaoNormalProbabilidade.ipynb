# Distribuição Normal - NERD DOS DADOS

```python
# Importação dos pacotes/bibliotecas
import pandas as pd
import seaborn as sns
from scipy.stats import norm
import numpy as np
import statistics as st
import matplotlib.pyplot as fig
```

```python
# Carregando o conjunto de dados
df = pd.read_csv('alturas.csv')
```

```python
# Visualizando as primeiras linhas

df.head()
```

```python
# Visualizando algumas medidas estatisticas
df.describe()
```

```python
# Valor mínimo

vlMin = df.min()
print(vlMin)
```

```python
# Valor máximo

vlMax = df.max()
print(vlMax)
```

```python
# Média

media = st.mean(df['altura'])
print(media)
```

```python
# Desvio padrão

desvio_padrao = st.pstdev(df['altura'])
print(desvio_padrao)
```

```python
# Histograma com a curva da distribuição
sns.histplot(df['altura'], kde = True, discrete = True)
```

```python
# Curva da distribuição
sns.kdeplot(data = df)
```

```python
# Gerando gráficos 
    
vlMin = df.min()
vlMax = df.max()
media = st.mean(df['altura'])
desvio_padrao = st.pstdev(df['altura'])
    
fig.subplot(211)
fig.plot(df['altura'], '-b')

fig.subplot(212)
fig.hist(df, bins = 10, density=True)

eix = np.linspace(vlMin,vlMax,100)
eiy = norm.pdf(eix,media,desvio_padrao)
fig.plot(eix, eiy, color = 'black')
```

```python
# Obtendo a probabilidade

probabilidade = norm.cdf(175,165,8) - norm.cdf(155,165,8)
round(probabilidade * 100,2)
```
