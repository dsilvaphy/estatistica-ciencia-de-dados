## CORRELAÇÃO

<P> Em probabilidade e estatística, correlação, dependência ou associação é qualquer relação estatística entre duas variáveis.<br>
    
<p>Resume o grau de relacionamento entre duas variáveis (X e Y, por exemplo). O objetivo do estudo da correlação é determinar (mensurar) o grau de relacionamento entre duas variáveis.<br>
    
<p>O valor do coeficiente de correlação pode variar entre -1 e 1 e o resultado obtido define se a correlação é negativa ou positiva. Para interpretar o coeficiente é preciso saber que 1 significa que a correlação entre as variáveis é perfeita positiva e -1 significa que é perfeita negativa. Se o coeficiente for igual a 0 significa que as variáveis não dependem uma da outra.    

<p>Importante: CORREÇÃO NÃO SIGNIFICA CAUSALIDADE<br>  


```python
# Um coeficiente de correlação próximo de zero indica que não há relação entre as duas variáveis
# E quanto mais eles se aproximam de 1 ou -1, mais forte é a relação.

from IPython.display import Image
Image('correlacao.png')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("dataset.xlsx")
```

```python
df.head(10)
```

```python
# Neste caso podemos dizer que a CORRELAÇÃO É NEUTRA OU NULA ou que Não existe CORRELAÇÃO entre essas variáveis.
sns.lmplot(x = "Ano", y = "Cilindradas", data = df);
```

```python
# Neste caso podemos dizer que existe uma CORRELAÇÃO FORTE POSITIVA entre essas variáveis
# Na medida que a variável X aumenta a variável Y também aumenta
sns.lmplot(x = "Ano", y = "ValorTabela", data = df);
```

```python
# Neste caso podemos dizer que existe uma CORRELAÇÃO FORTE NEGATIVA entre essas variáveis
# Na medida que a variável X aumenta a variável Y diminui
sns.lmplot(x = "Consumo_Litro_Km", y = "Cilindradas", data = df);
```

```python
# Aqui podemos observar os coeficientes da correção
df.corr()
```

```python
# Através do hitmap podemos observar a correlação entre todas variáveis.
plt.rcParams["figure.figsize"] = (18,8)
ax = sns.heatmap(df.corr(), annot=True) 
```

```python
sns.lmplot(x = "Cilindradas", y = "Peso_kg", data = df);
```
